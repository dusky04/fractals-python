```python
import numpy as np
from julia_set import JuliaSet
```

```python
def f(z, c: complex):
    return z * z + c


def f2(z, c: complex):
    return np.sin(z) + c
```

```python
J = JuliaSet(f=f, c=(-0.70176 - 0.3842j), height=1000, width=1000, max_iters=100)
```

```python
J.show_julia_set(cmap="Blues")
J.save_julia_set(fname="julia-set-1.png", cmap="Blues")
```

```python
J.c = 0 + 1j
J.show_julia_set(cmap="twilight")
J.save_julia_set(fname="julia-set-2.png", cmap="twilight")
```

```python
J.c = -0.8 + 0.156j
J.show_julia_set(cmap="twilight")
J.save_julia_set(fname="julia-set-3.png", cmap="twilight")
```
