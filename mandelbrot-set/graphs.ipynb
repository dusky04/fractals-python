```python
import numpy as np

from mandelbrot_set import MandelbrotSet
```

```python
def f1(z, c):
    return z * z + c


def f2(z, c):
    return z * np.sin(c)


def f3(z, c):
    return z - ((z - 1) ** 3 / (3 * z**2)) + c
```

```python
M = MandelbrotSet(f=f1, h=1000, w=1000, max_iters=100)

# default x_min = -2.0, y_min = -1.4, x_max = 0.8, y_max = 2.0
M.display_mandelbrot_set(cmap="cubehelix")
```

```python
M = MandelbrotSet(f=f2, h=1000, w=2000, max_iters=100)
M.display_mandelbrot_set(cmap="twilight_shifted", x_min=-2, y_min=-2, x_max=2, y_max=2)
```

```python
M = MandelbrotSet(f=f3, h=1000, w=2000, max_iters=100)
M.display_mandelbrot_set(cmap="cubehelix")
```
